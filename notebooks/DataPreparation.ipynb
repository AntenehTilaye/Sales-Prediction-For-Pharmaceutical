### Handle Missing Data

we can fill the following columns using mode, since both are date components
- CompetitionOpenSinceMonth
- CompetitionOpenSinceYear   

```python
column_names = ['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear']
store = du.fill_using_mode(store, column_names)
# the name and datatype of each column with null count
pd.DataFrame(data=list(zip(store.columns, store.dtypes, store.isna().sum(), round((store.isna().sum() * 100)/store.shape[0]))), columns=["name", "dtype", "# missing", "% Missing"])
```

we can fill the following column using median
- CompetitionDistance 

```python
column_names = ['CompetitionDistance']
store = du.fill_using_median(store, column_names)
# the name and datatype of each column with null count
pd.DataFrame(data=list(zip(store.columns, store.dtypes, store.isna().sum(), round((store.isna().sum() * 100)/store.shape[0]))), columns=["name", "dtype", "# missing", "% Missing"])
```

We can fill the following columns using 0, because all of them depend on the promo2, 
so when promo2 is 0, those columns values will be null
and when promo2 is 1, those columns will have values
###### we don't require a value there, since have no significance when promo2 is 0

- Promo2SinceWeek
- Promo2SinceYear
- PromoInterval

```python
store.fillna(0, inplace=True)
pd.DataFrame(data=list(zip(store.columns, store.dtypes, store.isna().sum(), round((store.isna().sum() * 100)/store.shape[0]))), columns=["name", "dtype", "# missing", "% Missing"])
```

- since the store is data is clean now, we can merge it the test and train data.

```python
merged_train = sales_train.merge(store, on='Store')
merged_train.head()
```

```python
merged_test = sales_test.merge(store, on='Store')
merged_test.head()
```
